```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

```

```python
data = pd.read_csv(r"C:\Users\PC\Desktop\Formation_Classifier\well_logs.csv")
```

## data.head()

```python
print(data[['GR', 'RT.OHMM', 'RXO.OHMM']].isnull().mean() * 100)
```

```python
data.replace(-999.25, np.nan, inplace=True)
```

```python
key_logs = ['GR', 'RT.OHMM', 'RXO.OHMM']
```

```python
data.dropna(subset=key_logs, how='all', inplace=True)
```

```python
data[key_logs] = data[key_logs].interpolate(method='linear')
```

```python
data.head(100)
```

```python
data.isnull().sum()
```

```python
def classify_formation(row):
    gr = row['GR']
    rt = row['RT.OHMM']
    
    if gr < 75 and rt > 50:
        return 'Clean Sandstone'
    elif gr > 150:
        return 'Shale'
    elif 75 <= gr <= 150:
        return 'Silty/Shaly Sand'
    else:
        return 'Unknown'
```

```python
data['Formation'] = data.apply(classify_formation, axis=1)
```

```python
data.info()
```

```python
data.reset_index()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.head(100)
```

```python
data.
```

```python
data["Formation"].value_counts()
```

```python
data_binary = data[data['Formation'].isin(['Clean Sandstone', 'Shale'])].copy()
```

```python
data_binary['Label']= data['Formation'].map({'Clean Sandstone':1, 'Shale':0})
```

```python
data_binary.info()
```

```python
data_binary['Formation'].value_counts()
```

```python
X = data_binary[['GR', 'RT.OHMM', 'RXO.OHMM']]  # or any other features you're using
y = data_binary['Formation']  # this should now be 0s and 1s
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size= 0.3, stratify=y, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)  # Make predictions on the test set
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

cv = StratifiedKFold(n_splits=5)
scores = cross_val_score(model, X, y, cv=cv)
print(scores.mean())  # Average accuracy across all 5 splits
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)  # 5-fold cross-validation
print("Cross-validation scores:", scores)
print("Average CV score:", scores.mean())
```

```python
import pandas as pd

feature_importance = pd.Series(model.feature_importances_, index=X.columns)
feature_importance.sort_values(ascending=False).plot(kind='bar')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

VISUALIZING THIS HERE TO SEE HOW IT LOOKS BEFORE PROCEEDING TO PETREL

```python
import matplotlib.pyplot as plt

def plot_well_logs(df, well_name = "GABO-14"):
    depth = df["DEPTH"]
    log_names = [col for col in df.columns if col!= 'DEPTH']

    num_logs = len(log_names)
    fig, axes = plt.subplots (nrows =1, ncols =num_logs, figsize = (15,15), sharey = True)

    for i, log_name in enumerate(log_names):
        if log_name in df.columns:
            ax = axes[i]
            ax.plot(df[log_name], depth)
            ax.set_xlabel(log_name)
            ax.invert_yaxis()
            ax.grid(True)

    axes[0].set_ylabel("Depth (M)")
    plt.suptitle(f"Well: {well_name} - All Logs")
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Call the plotting function with the dataframe
plot_well_logs(df)
```
